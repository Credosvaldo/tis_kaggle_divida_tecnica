## __Identification of Cassava Disease Symptoms__:




## Technical Instructions:
- Look for leaves drying and dying early.
- Check for angular spots on the leaves.
- Cut out small pieces of the leaf from the edge of the spots and place them in a drop of water.
- Observe for bacterial streaming - appears as white streaks in the water.
- Inspect for dark brown to black streaks on the green part of the stem.
- Check for the presence of sticky liquid.
- Look for browning in the vascular tissues after peeling the bark and splitting the stem.

## Explanation for Beginners:
- Check if the leaves are wilting or turning brown before they should.
- Look for odd-shaped spots with sharp corners on the leaves.
- Take a small piece of the leaf from the edges of these spots and put them in a drop of water.
- If you see white streaks in the water, it might indicate bacteria.
- Inspect the stem for dark lines, and if you find any, see if there's a sticky substance.
- Also, see if there's any sticky liquid present.
- Look for browning inside the stem's water tubes after removing its outer covering.

# <p style="color:#f90c0c;text-align:center;font-family:cursive;font-weight: bold;font-size:120%"> Evaluating Models </p>




```python
Batch_size = 16
img_height, img_width = 300, 300
```

```python
train_df['label'].dtype
```


# <p style="color:#f90c0c;text-align:center;font-family:cursive;font-weight: bold;font-size:120%"> Data Augmentation </p>



```python
train_df['label'] = train_df['label'].astype('str')
gen = ImageDataGenerator(
    horizontal_flip = True,
    vertical_flip = True,
    validation_split = 0.2,
)

train_datagen = gen.flow_from_dataframe(
    train_df,
    directory = os.path.join(Dir, "train_images"),
    batch_size = Batch_size,
    target_size = (img_height, img_width),
    subset = "training",
    seed = 42,
    x_col = "image_id",
    y_col = "label",
    class_mode = "categorical"
)
```

There are 17118 training images.

```python
val_gen = ImageDataGenerator(
    validation_split = 0.2
)

val_datagen = val_gen.flow_from_dataframe(
    train_df,
    directory = os.path.join(Dir, "train_images"),
    batch_size = Batch_size,
    target_size = (img_height, img_width),
    subset = "validation",
    seed = 42,
    x_col = "image_id",
    y_col = "label",
    class_mode = "categorical"
)
```

There are 4279 validation images.

```python
len(train_datagen), len(val_datagen)
```



When training a model with 21,397 images, we divide them into training and validation sets.

## For the training set:
- 80% of the images (21,397 * 0.8)
- Divide this by the batch size (16)
- Result: Training batch length is 1070, meaning the model processes 1070 batches of 16 images each.

## For the validation set:
- 20% of the images (21,397 * 0.2)
- Divide this by the batch size (16)
- Result: Validation batch length is 268, meaning the model processes 268 batches of 16 images each.

```python
Steps_per_train = float(train_datagen.n) / train_datagen.batch_size
Steps_per_val = float(val_datagen.n) / val_datagen.batch_size
```

```python
Steps_per_train, Steps_per_val
```

I have tried out various models like:

# <p style="color:#f90c0c;text-align:center;font-family:cursive;font-weight: bold;font-size:120%"> Build Model </p>





```python


def create_model():
    model = models.Sequential()
    model.add(EfficientNetB3(include_top=False, weights='imagenet',
                             input_shape=(img_height, img_width, 3), drop_connect_rate=0.3))
    model.add(layers.GlobalAveragePooling2D())
    model.add(layers.Flatten())
    model.add(layers.Dense(256, activation="relu"))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(5, activation='softmax'))
    
    loss = tf.keras.losses.CategoricalCrossentropy(
        label_smoothing=0.0001,
        name='categorical_crossentropy'
    )
    optimizer = optimizers.Adam(learning_rate=1e-4)
    
    model.compile(optimizer=optimizer,
                  loss=loss,
                  metrics=["categorical_accuracy"])
    return model

model = create_model()
model.summary()
```

```python
# Build the model
model.build((None))

# Plot the model
tf.keras.utils.plot_model(model, show_shapes=True, show_layer_names=True)
```

```python
rlronp=tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss",
                                            factor=0.2,
                                            mode = "min",
                                            min_lr=1e-6,
                                            patience=2, 
                                            verbose=1)

estop=tf.keras.callbacks.EarlyStopping(monitor="val_loss", 
                                       mode= "min",
                                       patience=3, 
                                       verbose=1,
                                       restore_best_weights=True)

history = model.fit(
    train_datagen,
    steps_per_epoch=int(Steps_per_train),
    epochs=5,
    verbose =1,
    validation_data=val_datagen,
    validation_steps=int(Steps_per_val),
    callbacks=[rlronp, estop]
)

```

```python

train_acc = history.history["categorical_accuracy"]
val_acc = history.history["val_categorical_accuracy"]
epochs = range(1, len(train_acc)+1)
plt.plot(epochs, train_acc, "bo", label = "Training Accuracy")
plt.plot(epochs, val_acc, "b", label = "Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()
```

```python

plt.figure(figsize=(8,6))
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
epochs = range(1, len(train_loss)+1)
plt.plot(epochs, train_loss, "bo", label = "Training Loss")
plt.plot(epochs, val_loss, "b", label = "Validation Loss")
plt.title("Training and Validation Loss")
plt.legend()
```
